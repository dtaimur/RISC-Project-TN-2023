## Normalized Frequency for one provider ID

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert the timestamp column to datetime and extract the hour and day of the week
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['hour'] = df['Timestamp'].dt.hour
df['day_of_week'] = df['Timestamp'].dt.dayofweek  # Monday is 0 and Sunday is 6

# Define a function to calculate the normalized hourly frequency for a specific day
def calculate_normalized_hourly_frequency(data):
    hourly_frequency = data['hour'].value_counts().sort_index()
    mean_hourly_frequency = hourly_frequency.mean()
    std_hourly_frequency = hourly_frequency.std()
    normalized_hourly_frequency = (hourly_frequency - mean_hourly_frequency) / std_hourly_frequency
    return normalized_hourly_frequency

# Set the Provider ID you want to compare
provider_id = '3fd53b10-98a7-3a86-9830-19b34b61d27f'

# Filter the dataset for the specific Provider ID
provider_data = df[df['Provider ID'] == provider_id]

# Create a dictionary to store the normalized hourly frequency for each day of the week
normalized_hourly_frequency_per_day = {}

# Calculate the normalized hourly frequency for each day of the week
for day in range(7):
    day_data = provider_data[provider_data['day_of_week'] == day]
    normalized_hourly_frequency_per_day[day] = calculate_normalized_hourly_frequency(day_data)

# Calculate the whole dataset line for each day of the week
normalized_hourly_frequency_whole_per_day = {}
for day in range(7):
    day_data = df[df['day_of_week'] == day]
    normalized_hourly_frequency_whole_per_day[day] = calculate_normalized_hourly_frequency(day_data)

# Plot the normalized hourly frequency for each day and the whole dataset line
plt.figure(figsize=(12, 15))
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

for day in range(7):
    plt.subplot(4, 2, day + 1)

    # Plot the Provider ID line for this day
    if not normalized_hourly_frequency_per_day[day].empty:
        plt.plot(normalized_hourly_frequency_per_day[day].index, normalized_hourly_frequency_per_day[day],
                 label=f'Provider ID: {provider_id}', color='orange')

    # Plot the whole dataset line for this day
    plt.plot(normalized_hourly_frequency_whole_per_day[day].index, normalized_hourly_frequency_whole_per_day[day],
             label='Whole Dataset', linewidth=2, color='blue')

    plt.axhline(y=3, color='r', linestyle='--', label='Anomalous Threshold')
    plt.axhline(y=-3, color='r', linestyle='--')
    plt.title(f'Normalized Hourly Frequency of Activities - {days_of_week[day]}')
    plt.xlabel('Hour')
    plt.ylabel('Normalized Frequency')
    plt.legend()

    # Set the x-axis limit to have the same scale for all graphs (hours 0 to 23)
    plt.xlim(0, 23)

plt.tight_layout()
plt.show()
```

##Normalized Frequency, anomaly detection for the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert the timestamp column to datetime and extract the hour and day of the week
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['hour'] = df['Timestamp'].dt.hour
df['day_of_week'] = df['Timestamp'].dt.dayofweek  # Monday is 0 and Sunday is 6

# Calculate the mean and standard deviation of the hourly frequency for the whole dataset
mean_hourly_frequency_whole = df['hour'].value_counts().sort_index().mean()
std_hourly_frequency_whole = df['hour'].value_counts().sort_index().std()

# Calculate the normalized hourly frequency for the whole dataset for each day of the week
normalized_hourly_frequency_whole_per_day = {}
for day in range(7):
    hourly_frequency_day = df[df['day_of_week'] == day]['hour'].value_counts().sort_index()
    normalized_hourly_frequency_day = (hourly_frequency_day - mean_hourly_frequency_whole) / std_hourly_frequency_whole
    normalized_hourly_frequency_whole_per_day[day] = normalized_hourly_frequency_day

# Define the anomaly threshold
anomalous_threshold = 3

# Iterate through all unique provider IDs
for provider_id in df['Provider ID'].unique():
    # Filter the dataset for the specific Provider ID
    provider_data = df[df['Provider ID'] == provider_id]

    # Create an empty dictionary to store the normalized hourly frequency for each day of the week
    normalized_hourly_frequency_per_day = {}

    # Calculate the mean and standard deviation of the hourly frequency for each day of the week
    for day in range(7):
        day_data = provider_data[provider_data['day_of_week'] == day]
        hourly_frequency_day = day_data['hour'].value_counts().sort_index()
        mean_hourly_frequency_day = hourly_frequency_day.mean()
        std_hourly_frequency_day = hourly_frequency_day.std()
        normalized_hourly_frequency_day = (hourly_frequency_day - mean_hourly_frequency_day) / std_hourly_frequency_day
        normalized_hourly_frequency_per_day[day] = normalized_hourly_frequency_day

    # Check if the normalized hourly frequency for any day falls outside the anomaly threshold
    is_anomalous = False
    anomalous_day = None
    for day, normalized_hourly_frequency_day in normalized_hourly_frequency_per_day.items():
        if (normalized_hourly_frequency_day > anomalous_threshold).any() or (normalized_hourly_frequency_day < -anomalous_threshold).any():
            is_anomalous = True
            anomalous_day = day
            break

    # If the provider is anomalous, plot its graph with the whole dataset's normalized hourly frequency for that specific day of the week
    if is_anomalous:
        plt.figure(figsize=(12, 6))
        day_of_week_label = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
        plt.plot(normalized_hourly_frequency_whole_per_day[anomalous_day].index, normalized_hourly_frequency_whole_per_day[anomalous_day], label=f'Whole Dataset - {day_of_week_label[anomalous_day]}', color='blue', alpha=0.5)
        plt.plot(normalized_hourly_frequency_per_day[anomalous_day].index, normalized_hourly_frequency_per_day[anomalous_day], label=f'Provider ID: {provider_id} - {day_of_week_label[anomalous_day]}', color='orange')
        plt.axhline(y=anomalous_threshold, color='r', linestyle='--', label='Anomalous Threshold')
        plt.axhline(y=-anomalous_threshold, color='r', linestyle='--')
        plt.title(f'Normalized Hourly Frequency of Activities')
        plt.xlabel('Hour')
        plt.ylabel('Normalized Frequency')
        plt.legend()
        plt.show()
```

#Signal Flags

##Signal Flag list and frequency

```python
def detect_flag_anomalies(df, threshold):
    # Count the number of occurrences for each signal flag
    flag_counts = df['Signal Flags'].value_counts()

    # Identify signal flags that appear more frequently than the threshold
    flag_anomalies = flag_counts[flag_counts > threshold]

    return flag_anomalies

# Set the threshold for flag-based anomalies
flag_threshold = 0  # adjust the threshold based on your preference

# Detect flag-based anomalies in the dataset
flag_anomalies = detect_flag_anomalies(df, flag_threshold)

# Print the flag-based anomalies
print("Flag-based Anomalies:")
print(flag_anomalies)
```
